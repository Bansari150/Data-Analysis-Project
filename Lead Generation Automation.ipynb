```python
import pandas as pd

def run_automation():

    df = pd.read_csv("companies.csv")

    df.drop_duplicates(inplace=True)

    df.fillna("Not Available", inplace=True)

    df["Generated_Email"] = (
        df["Company_Name"]
        .str.lower()
        .str.replace(" ", "")
        + "@gmail.com"
    )

    df.to_excel("cleaned_companies.xlsx", index=False)

    print("Automation completed successfully!")

run_automation()
from google.colab import files
files.download("cleaned_companies.xlsx")
print("Excel file created successfully!")
```
